```python
#get filename under some data file
pathdir = '/data/xueyou/car/car_news_0606'
filenames = []
for alldir in os.listdir(pathdir):
    if alldir != "README.txt":
        #child = os.path.join('%s/%s' % (pathdir, alldir))
        child = alldir
        filenames.append(child)
print(filenames)

#1、将每一个file文件中的行文本数据处理为dict的格式
#2、将处理完的dict格式汇总为list数据结构
def str_to_dict(str1):
    dict = {}
    dict['article_type'] = str1[0]
    dict['brand'] = str1[1]
    dict['serial'] = str1[2]
    dict['author'] = str1[3]
    dict['headline'] = str1[4]
    dict['content'] = str1[5]
    dict['pubdate'] = str1[6]
    return dict
    
def file_to_list(file):
    res = []
    for line in file.readlines():
        line1 = line.split('\x01')
        if len(line1) < 7:
            continue
        article = str_to_dict(line1)
        res.append(article)
    return res

#get all articles
articles = []
for path in filenames:
    path1 = pathdir + '/' + path
    #print(path1)
    for child in os.listdir(path1):
        #print(child.split('.'))
        if 'crc' not in child.split('.'):
            #print(child)
                #child1 = path + '/' + child
                #print(child1)
            file = open(path + '/'+child, 'r', encoding='utf-8') 
            articles0 = file_to_list(file)
            articles.extend(articles0)   
```

```python
len(articles)
```

```python
articles0 = []
for art in articles:
    if art in articles0:
        continue
    else:
        articles0.append(art)
```

```python
len(articles0)
```

```python
#‘pubdate’处理为datetime格式数据
def str_to_date(pubdate):
    date1 = datetime.datetime.strptime(pubdate.split('T')[0], "%Y-%m-%d").date()
    return date1

def need_cluster(article, cluster_article):    #cluster_article 是一篇文章
    if not article['serial']:                  #判断文章的车系信息是否为空，没有车系信息，就不对他聚类
        return False

    #如果车系信息，品牌信息与文章聚类好的信息对应， 文章类行信息不对应与“其他”，返回True
    if article['serial'] == cluster_article['serial'] and \
        article['brand'] == cluster_article['brand'] and \
        article['article_type'] != '其他'and \
        article['article_type'] != '\\N' and \
        article['article_type'] == cluster_article['article_type'] and \
        abs((str_to_date(article['pubdate']) - str_to_date(cluster_article['pubdate'])).days) <= 60 :
        return True
    else:
        return False

def text_cluster(articles):     #文本聚类（类别）list
    cluster_list = []
    for article in articles:
        new_cluster_flag = True
        for cluster_article in cluster_list:
            if need_cluster(article, cluster_article):              #如果article与cluster_article的车辆信息或关键词匹配
                cluster_article["related_articles"].append(article) #则将article作为cluster_article的关联文章
                new_cluster_flag = False
                break
        if new_cluster_flag:                 #如果没有与article相匹配的cluster_article
            if article['article_type'] != '其他'and article['article_type'] != '\\N':
                article["related_articles"] = [] #则为article的‘related’参数设为空列表，并将这一文章添加进cluster_list
                cluster_list.append(article)
    return cluster_list                      #返回按照文章内容中具有相似车辆信息的文章
```

```python
cluster_list = text_cluster(articles0)
```

```python
len(cluster_list)
```

```python
#下面在articles0的基础上，对cluster_list进行list化
cluster_lists = []
for articles_1 in cluster_list:
    #print(article_list)
    list_articles = []
    related_article = articles_1['related_articles']
    for article in related_article:
        if article['content'][:30] == articles_1['content'][:30]:
            continue
        else:
            list_articles.append(article)
        #else:
        #   continue
    del articles_1['related_articles']
    list_articles.append(articles_1)
    cluster_lists.append(list_articles)
```

```python
#再每一个cluster内过滤相似的文本
def cluster_filter(act_list):
    res_list = []
    for act in act_list:
        if act in res_list:
            continue
        else:
            res_list.append(act)
    return res_list

def filted_cluster(clu):
    cluster_filted = []
    for a in clu:
        cluster_filted.append(cluster_filter(a))
    return cluster_filted
```

```python
cluster_final = filted_cluster(cluster_lists)
```
